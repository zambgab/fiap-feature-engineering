> ### Desenvolvimento do baseline sem feature engineering robusto ou uso de variáveis categóricas

- Separação das variáveis quantitativas
- Inputação de dados faltantes com uso da mediana das colunas
- Treinamento do modelo baseline
- Avaliação do modelo para conhecimento do pior caso

> #### Separação das variáveis quantitativas

```python
# Seleciona as variáveis quantitativas
x = raw[["QTD_CARTOES_ADICIONAIS", "QTD_DEPENDENTES", "MESES_RESIDENCIA", "RENDA_PESSOAL_MENSAL", "OUTRAS_RENDAS", "QTD_CONTAS_BANCARIAS", 
                  "QTD_CONTAS_BANCARIAS_ESPECIAIS", "VLR_PATRIMONIO_PESSOAL", "QTD_CARROS"]]

# Seleciona a variável target
y = raw["TARGET"]
```

> #### Inputação de dados faltantes com uso da mediana das colunas

```python
# Verifica a presença de valores nulos nas variáveis quantitativas
x.isnull().sum()
```

```python
# Gera a mediana da variável que possui valores nulos
mediana_meses_residencia = x["MESES_RESIDENCIA"].median()

# Utiliza a mediana da variável para preencher os valores nulos
x.loc[:, "MESES_RESIDENCIA"] = x["MESES_RESIDENCIA"].fillna(mediana_meses_residencia)

# Verifica a existência de valores nulos
x.isnull().sum()
```

```python
# Verifica a presença de valores nulos na variável target
y.isnull().sum()
```

> #### Treinamento do modelo baseline

```python
# Divide o conjunto de dados em treino (70%) e teste (30%)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

# Define o classificador KNN como modelo a ser treinado
knn = KNeighborsClassifier()

# Gera uma lista com valores de 1 a 25 para testar o melhor valor de K
k_values = {"n_neighbors" : [i for i in range(1,26)]}

# Uso do método GridSearchCV para realizar uma busca exaustiva pelo melhor valor de K
grid_search = GridSearchCV(knn, k_values, cv=5, scoring="accuracy")

# Treina o modelo com o melhor valor de K
grid_search.fit(x_train, y_train)

# Prevê com os dados de teste
y_pred = grid_search.predict(x_test)

# Apresenta o melhor valor de K
best_k = grid_search.best_params_["n_neighbors"]
print(f"O melhor valor para K é {best_k}\n")

# Apresenta o relatório de classificação do modelo
report = classification_report(y_test, y_pred)
print(f"Relatório de classificação do modelo\n\n{report}\n")

# Apresenta a matriz de confusão do modelo
conf_matrix = confusion_matrix(y_test, y_pred)
print(f"Matriz de confusão\n\n{conf_matrix}\n")
```

Note que, mesmo após a exclusão das variáveis com valores nulos ou sem variabilidade, ainda existem registros vazios no conjunto de dados, como apresentado na variável sexo - denotado por " ". Dessa forma, será analisada a presença de dados ausentes em cada uma das características.

Sendo assim, a função "gera_histograma" será criada para que seja mais fácil identificar a presença de sujeiras na conjunto de dados, o que permitirá uma atuação mais precisa na correção das variáveis.

```python
# Cria a função para gerar histogramas
def gera_histograma(feature, n_bins):
    """
    Entrada: a função recebe dois parâmetros: feature, ou característica a ser analisada; n_bins, ou números de barras (valores possíveis da feature em questão).
    Saída: a função apresenta o histograma da variável de entrada.
    """
    # Gera área do gráfico a ser apresentado
    plt.figure(figsize=(16,4))

    # Plota o histograma da variável
    n, bins, patches = plt.hist(stage[f"{feature}"], bins=n_bins, edgecolor="black", alpha=0.7)

    # Adiciona o número de ocorrências para cada valor da variável
    for count, x in zip(n, bins):
        height = count 
        plt.text(x + (bins[1] - bins[0]) / 2, height, int(count), ha="center", va="bottom")

    # Adiciona título ao gráfico
    plt.title(f"Distribuição dos dados para variável {feature}")

    # Adiciona rótulo ao eixo X
    plt.xlabel("Variável")

    # Adiciona rótulo ao eixo Y
    plt.ylabel("Frequência")

    # Exibe o histograma
    plt.show()
```

```python
# Cria a função para normalizar variáveis de texto
def normaliza_texto(feature):
    """
    Entrada: a função recebe um parâmetro: feature, ou característica a ser analisada.
    Saída: a função retorna a feature normalizada, ou seja, com todas os tratamentos listados abaixo:
        - Letras maiúsculas
        - Sem espaços a direita ou a esquerda
        - Sem caractéres numéricos (são substituídos por "")
        - Sem n-espaços ("  ", "   ", "    ", "     " são substituídos por " ")
        - Sem caractéres especiais ("$", por exemplo, é substituído por "")
    """
    # Apresenta quantos valores únicos a variável possui pré normalização
    unique_values = len(stage[f"{feature}"].unique())
    print(f"A variável {feature} possui {unique_values} valores únicos antes da normalização.\n")
    
    # Converte os valores da variável para maiúsculo
    stage[f"{feature}"] = stage[f"{feature}"].str.upper()

    # Retira espaços a direita e a esquerda dos valores da variável
    stage[f"{feature}"] = stage[f"{feature}"].str.strip()
    
    # Substitui espaços n-espaços (2 a 5) por espaços simples
    stage[f"{feature}"] = stage[f"{feature}"].str.replace("  ", " ").replace("   ", " ").replace("    ", " ").replace("     ", " ")

    # Cria um lista com o conteúdo da variável sem tratamentos
    words = list(stage[f"{feature}"])

    # Cria uma lista que armazena o conteúdo da variável pós tratamentos
    wordsNormalized = list()

    # Percorre todas as palavras da lista de palavras
    for word in words:
        # Cria uma palavra temporária vazia
        temporaryWord = ""
        # Percorre todos os caractéres da i-ésima palavra
        for char in range(len(word)):
            # Junta os caractéres da i-ésima palavra para formar a palavra completa
            temporaryWord += word[char].replace("0", "").replace("1", "").replace("2", "").replace("3", "").replace("4", "").replace("5", "").replace("6", "").replace("7", "").replace("8", "").replace("9", "").replace("Á", "A").replace("Ã", "A").replace("À", "A").replace("Â", "A").replace("É", "E").replace("Ê", "E").replace("È", "E").replace("Í", "I").replace("Î", "I").replace("Ì", "I").replace("Ó", "O").replace("Ô", "O").replace("Ò", "O").replace("Õ", "O").replace("Ú", "U").replace("Ù", "U").replace("!", "").replace("@", "").replace("#", "").replace("$", "").replace("%", "").replace("¨", "").replace("&", "").replace("*", "").replace("(", "").replace(")", "").replace("-", "").replace("_", "").replace("=", "").replace("+", "").replace("'", "").replace("´", "").replace("`", "").replace("[", "").replace("{", "").replace("]", "").replace("}", "").replace("Ç", "C").replace("~", "").replace("^", "").replace(",", "").replace("<", "").replace(".", "").replace(">", "").replace(";", "").replace(":", "")
        # Adiciona a i-ésima palavra a lista de palavras normalizadas
        wordsNormalized.append(temporaryWord)
    
    # Cria uma Series com base na lista de palavras normalizadas
    featureNormalized = pd.Series(wordsNormalized, name=f"{feature}")
    
    # Apresenta quantos valores únicos a variável possui pós normalização
    uniqueValues = len(featureNormalized.unique())
    print(f"A variável {feature} possui {uniqueValues} valores únicos depois da normalização.")

    return featureNormalized
```

De qualquer forma, visto a grande variabilidade entre os valores máximos e mínimos da variável MESES_RESIDENCIA, será realizada a remoção de outliers para diminuir a possibilidade de classificações incorretas.

```python
# Cria a função para detectar e excluir outliers
def detectar_outliers(dataset, feature):
    """
    Entrada: a função recebe dois parâmetros: dataset, ou conjunto de dados; feature, ou característica a ser analisada.
    Saída: a função retorna o número de registros antes da remoção de outliers; número de outliers; número de registros depois da remoção de outliers.
    """
    # Calcula os quartis
    q1 = np.percentile(dataset[f"{feature}"], 25)
    q3 = np.percentile(dataset[f"{feature}"], 75)
    
    # Calcula o intervalo interquartil
    iqr = q3 - q1
    
    # Define limites para detectar outliers
    lower_fence = q1 - 1.5 * iqr
    upper_fence = q3 + 1.5 * iqr

    # Retorna o limite superior e inferior
    return [float(lower_fence), float(upper_fence)]
```

> #### Para determinar se existe uma associação significativa entre duas variáveis categóricas (teste de independência), utilizaremos o teste do qui-quadrado, que verifica se a distribuição observada das frequências em categorias é diferente da distribuição esperada, assumindo que as variáveis sejam independentes. Temos as seguintes hipóteses do teste:
> #### Hipótese nula (H0): As variáveis são independentes, ou seja, não há associação entre elas.
> #### Hipótese alternativa (H1): As variáveis são dependentes, ou seja, há uma associação entre elas.

> #### Qui-quadrado (chi2): O valor calculado do qui-quadrado. Quanto maior esse valor, maior a diferença entre os dados observados e esperados, indicando uma possível associação entre as variáveis.

> #### P-valor (p): O valor que indica a probabilidade de observar os resultados obtidos se a hipótese nula for verdadeira. Um p-valor pequeno (geralmente menor que 0,05) sugere que devemos rejeitar a hipótese nula e aceitar que há uma associação significativa entre as variáveis.

> #### Graus de liberdade (dof): Representa o número de valores independentes que podem variar na tabela de contingência, calculado como (n_linhas - 1) * (n_colunas - 1).

> #### Frequências esperadas (expected): As frequências que seriam esperadas para cada célula da tabela se as variáveis fossem independentes.

```python
# Seleciona as variáveis quantitativas
x = stage[["QTD_DEPENDENTES", "MESES_RESIDENCIA" ,"RENDA_TOTAL_MENSAL", "QTD_TOTAL_CONTAS_BANCARIAS", "QTD_CARROS"]]

# Seleciona a variável target
y = stage["TARGET"]

# Seleciona as variáveis qualitativas (categóricas)
z = stage[["DIA_PAGAMENTO", "TIPO_ENVIO_APLICACAO", "SEXO", "ESTADO_CIVIL", "ESTADO_NASCIMENTO", 
           "ESTADO_RESIDENCIAL", "FLAG_TELEFONE_RESIDENCIAL","TIPO_RESIDENCIA", "FLAG_EMAIL", 
           "EMPRESA", "FLAG_TELEFONE_PROFISSIONAL", "PRODUTO", "IDADE", "CEP_RESIDENCIAL_3", "FAIXA_RENDA_TOTAL_MENSAL", "FLAG_BANDEIRAS"]]
```

```python
'''
Vamos calcular o qui-quadrado para cada uma das variáveis categóricas do dataset em relação à variável TARGET 
para verificar se há uma associação significativa entre cada variável e a variável target
'''

# Inicializar uma lista vazia para armazenar os resultados
chi2_results = []

# Realizar o teste do qui-quadrado para cada variável categórica do dataset
for column in z:
    # Criar uma tabela de contingência entre a variável categórica e a TARGET
    contingency_table = pd.crosstab(z[column], stage['TARGET'])

    # Executar o teste do qui-quadrado
    chi2, p, dof, expected = chi2_contingency(contingency_table)

    # Armazenar os resultados em uma lista
    chi2_results.append({'Variável': column, 'Qui-Quadrado': chi2, 'P-valor': p, 'Graus de Liberdade': dof})

# Converter os resultados para um DataFrame para melhor visualização
chi2_results_df = pd.DataFrame(chi2_results)

# Exibir os resultados
chi2_results_df.sort_values(by='P-valor')
```

```python
# Interpretando o resultado
alpha = 0.05

'''
Criando a coluna hipótese que diz se rejeitamos a hipótese nula (há uma associação significativa entre as variáveis) 
ou se não rejeitamos (não há evidência suficiente para afirmar uma associação)
'''
chi2_results_df['Hipotese'] = chi2_results_df['P-valor'].apply(lambda p: "Há associação significativa" if p < alpha
                                                               else "Não há evidência suficiente para afirmar associação")

# Exibir os resultados
chi2_results_df.sort_values(by='P-valor')
```
